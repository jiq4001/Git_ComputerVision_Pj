```python
# Compute harris values of an image.
def computeHarrisValues(srcImage):
        '''
        Input:
            srcImage -- Grayscale input image in a numpy array with
                        values in [0, 1]. The dimensions are (rows, cols).
        Output:
            harrisImage -- numpy array containing the Harris score at
                           each pixel.
            orientationImage -- numpy array containing the orientation of the
                                gradient at each pixel in degrees.
        '''
        
        height, width = srcImage.shape[:2]

        harrisImage = np.zeros(srcImage.shape[:2])
        orientationImage = np.zeros(srcImage.shape[:2])
        
        # identity gaussian
        idf_gauss = np.zeros((5,5))
        idf_gauss[2,2] = 1
        idf_gauss_s05 = gaussian_filter(idf_gauss, sigma = 0.5, mode='reflect')
        
        # sobel of gaussian
        dx = sobel(idf_gauss_s05, axis = 0, mode = 'reflect')
        dy = sobel(idf_gauss_s05, axis = 1, mode = 'reflect')
        
        # convolution to image
        img_dx = convolve(srcImage, dx, mode = 'reflect')
        img_dy = convolve(srcImage, dy, mode = 'reflect')
        
        for i in range(height):
            for j in range(width):
                h11 = np.sum(img_dx[i, j] * img_dx[i, j])
                h12 = h21 = np.sum(img_dx[i, j] * img_dy[i, j])
                h22 = np.sum(img_dy[i, j] * img_dy[i, j])
                H = np.array([[h11, h12], [h21, h22]])
                harrisImage[i, j] = np.linalg.det(H) - 0.1 * np.trace(H) * np.trace(H)
                #zero angle points to the right and positive angles are counter-clockwise.
                if (img_dx[i, j] == 0) & (img_dy[i, j] == 0):
                    orientationImage[i, j] = 0
                elif (img_dx[i, j] == 0) & (img_dy[i, j] < 0):
                    orientationImage[i, j] = 270
                elif (img_dx[i, j] == 0) & (img_dy[i, j] > 0):
                    orientationImage[i, j] = 90
                elif (img_dx[i, j] > 0) & (img_dy[i, j] > 0):
                    orientationImage[i, j] = math.degrees(math.atan(img_dy[i, j]/img_dx[i, j]))
                elif (img_dx[i, j] < 0) & (img_dy[i, j] > 0):
                    orientationImage[i, j] = 90 - math.degrees(math.atan(img_dy[i, j]/img_dx[i, j]))
                elif (img_dx[i, j] < 0) & (img_dy[i, j] < 0):
                    orientationImage[i, j] = 180 + math.degrees(math.atan(img_dy[i, j]/img_dx[i, j]))
                else : #(img_dx[i, j] > 0) & (img_dy[i, j] < 0)
                    orientationImage[i, j] = 270 - math.degrees(math.atan(img_dy[i, j]/img_dx[i, j])) ### Angle or arc??? 

        # TODO 1: Compute the harris corner strength for 'srcImage' at
        # each pixel and store in 'harrisImage'.  See the project page
        # for direction on how to do this. Also compute an orientation
        # for each pixel and store it in 'orientationImage.'
        # TODO-BLOCK-BEGIN
        #raise Exception("TODO 1: in features.py not implemented")
        # TODO-BLOCK-END

        return harrisImage, orientationImage
```
